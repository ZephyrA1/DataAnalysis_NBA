### Maximum data points
The charts compares key performance Maximum value metrics for three basketball legends during a Regular and Playoff NBA season. At the peak of their career LeBron James, Michael Jordan, and Kobe Bryant each played 82 games, with Michael Jordan leading in points per game, followed by Kobe Bryant and LeBron James in the Regular and Playoff Season . The chart also highlights their shooting percentages, rebounding, assists, steals, and blocks, showcasing their diverse skill sets on the court.

Maximum values performed by each player in Both Regular and Playoff Season.

```python
# Maximum values performed by each player for each season
GOAT = pd.DataFrame(GOAT)

# Split the DataFrame into 'Regular Season' and 'Playoffs'
regular_season_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season']
playoffs_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs']

# Specify the desired columns
desired_columns = ['Games Played', 'Field Goal Percentage', 'Three Point Percentage', 'Two Point Percentage', 'Free Throws Percentage', 'Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create a function to calculate max values
def calculate_max_values(data):
    max_values = {option: {column: data[data['Player Name'] == option][column].max() for column in desired_columns} for option in data['Player Name'].unique()}
    return pd.DataFrame(max_values).transpose()

# Calculate max values for 'Regular Season' and 'Playoffs'
max_values_regular_season = calculate_max_values(regular_season_data)
max_values_playoffs = calculate_max_values(playoffs_data)

# Print the tables using tabulate
print("\nMax Values for Regular Season:")
print(tabulate(max_values_regular_season, headers='keys', tablefmt='grid'))

print("\nMax Values for Playoffs:")
print(tabulate(max_values_playoffs, headers='keys', tablefmt='grid'))
```

## Maximum Values 
### Performed Overall for Both Playoff and Regular Season

The chart summarizes the overall maximum statistical values for various basketball performance metrics during the regular season and playoffs. LeBron James holds the highest field goal percentage (0.567), played the most games (82), and leads in two-point percentage (0.622), total rebound percentage (13.1), total assist percentage (44.4), and total points per game in the regular season (37.1). In the playoffs, LeBron James also excels in several categories, including field goal percentage (0.565), games played (23), two-point percentage (0.625), and total rebound percentage (14.7), while Michael Jordan dominates in three-point percentage (1), total steal percentage (3.5), total block percentage (3.4), and total points per game (43.7).

```python
# Overall maximum values performed for each season
GOAT = pd.DataFrame(GOAT)

# Specify the desired columns
desired_columns = ['Field Goal Percentage', 'Games Played', 'Three Point Percentage', 'Two Point Percentage', 'Free Throws Percentage', 'Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create dictionaries to store overall max values and corresponding names for each desired column
overall_max_values_regular_season = {column: {'value': None, 'name': None} for column in desired_columns}
overall_max_values_playoffs = {column: {'value': None, 'name': None} for column in desired_columns}

# Iterate over each desired column
for column in desired_columns:
    # Find the overall max value and corresponding name for Regular Season
    overall_max_name_regular_season = GOAT.loc[GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season'][column].idxmax(), 'Player Name']
    overall_max_value_regular_season = GOAT.loc[GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season'][column].idxmax(), column]
    
    # Update the dictionary for Regular Season
    overall_max_values_regular_season[column]['value'] = overall_max_value_regular_season
    overall_max_values_regular_season[column]['name'] = overall_max_name_regular_season

    # Find the overall max value and corresponding name for Playoffs
    overall_max_name_playoffs = GOAT.loc[GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs'][column].idxmax(), 'Player Name']
    overall_max_value_playoffs = GOAT.loc[GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs'][column].idxmax(), column]
    
    # Update the dictionary for Playoffs
    overall_max_values_playoffs[column]['value'] = overall_max_value_playoffs
    overall_max_values_playoffs[column]['name'] = overall_max_name_playoffs

# Create DataFrames from the results for Regular Season and Playoffs
result_df_regular_season = pd.DataFrame(overall_max_values_regular_season).transpose()
result_df_playoffs = pd.DataFrame(overall_max_values_playoffs).transpose()

# Print the tables using tabulate
print("\nOverall Max Values for Regular Season:")
print(tabulate(result_df_regular_season, headers='keys', tablefmt='grid'))

print("\nOverall Max Values for Playoffs:")
print(tabulate(result_df_playoffs, headers='keys', tablefmt='grid'))
```

## Minimum data points

### Minimum values performed by each player for Regular and Playoff Season

The chart provides minimum values for various basketball statistics during the regular season and playoffs for three legendary players: Kobe Bryant, LeBron James, and Michael Jordan. In the regular season, Kobe Bryant had the lowest field goal percentage at 35.8%, LeBron James had the lowest three-point percentage at 29%, and Michael Jordan had the lowest two-point percentage at 40.3%. In the playoffs, Kobe Bryant again had the lowest field goal percentage at 38.2%, LeBron James had the lowest three-point percentage at 22.7%, and Michael Jordan had the lowest two-point percentage at 41.8%.

```python
# Minimum values performed by each player for each season
GOAT = pd.DataFrame(GOAT)

# Split the DataFrame into 'Regular Season' and 'Playoffs'
regular_season_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season']
playoffs_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs']

# Specify the desired columns
desired_columns = ['Games Played', 'Field Goal Percentage', 'Three Point Percentage', 'Two Point Percentage', 'Free Throws Percentage', 'Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create a function to calculate min values
def calculate_min_values(data):
    min_values = {column: data[column].min() if column != 'Games Played' else data[column].max() for column in desired_columns}
    return pd.DataFrame([min_values])

# Calculate min values for 'Regular Season' and 'Playoffs' for each player
min_values_regular_season = regular_season_data.groupby('Player Name').apply(calculate_min_values).reset_index()
min_values_playoffs = playoffs_data.groupby('Player Name').apply(calculate_min_values).reset_index()

# Print the tables using tabulate
print("\nMin Values for Regular Season:")
print(tabulate(min_values_regular_season, headers='keys', tablefmt='grid', showindex=False))

print("\nMin Values for Playoffs:")
print(tabulate(min_values_playoffs, headers='keys', tablefmt='grid', showindex=False))
```

## Overall minimum values performed by each player for Regular and Playoff Season

The chart provides the overall minimum values for various basketball performance metrics in both the regular season and playoffs. Kobe Bryant holds the lowest regular season records in Field Goal Percentage (0.358), Games Played (6), Two Point Percentage (0.401), Total Rebound Percentage (6.7), Total Assist Percentage (13.8), Total Block Percentage (0.3), and Total Points Per Game (7.6). In playoffs, Kobe Bryant maintains the lowest values in Field Goal Percentage (0.382), Total Rebound Percentage (5.1), Total Assist Percentage (12.1), Total Steal Percentage (0.7), Total Block Percentage (0.1), and Total Points Per Game (8.2).

```python
# Overall minimum values performed for each season
GOAT = pd.DataFrame(GOAT)

# Split the DataFrame into 'Regular Season' and 'Playoffs'
regular_season_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season']
playoffs_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs']

# Specify the desired columns
desired_columns = ['Field Goal Percentage', 'Games Played', 'Three Point Percentage', 'Two Point Percentage', 'Free Throws Percentage', 'Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create a function to calculate overall min values
def calculate_overall_min_values(data):
    overall_min_values = {column: {'value': data[column].min(), 'name': data.loc[data[column].idxmin(), 'Player Name']} for column in desired_columns}
    return pd.DataFrame(overall_min_values).transpose()

# Calculate overall min values for 'Regular Season' and 'Playoffs'
overall_min_values_regular_season = calculate_overall_min_values(regular_season_data)
overall_min_values_playoffs = calculate_overall_min_values(playoffs_data)

# Print the tables using tabulate
print("\nOverall Min Values for Regular Season:")
print(tabulate(overall_min_values_regular_season, headers='keys', tablefmt='grid'))

print("\nOverall Min Values for Playoffs:")
print(tabulate(overall_min_values_playoffs, headers='keys', tablefmt='grid'))
```

###  Regular Season and Playoffs Graphs Based on Best Perfomance by Player

The highest possible values for rebound %, assist percentage, steal percentage, block percentage, and points per game are extracted after filtered using the information for each players (Lebron James, Michael Jordan, and Kobe Bryant) and then is applied. A grouped bar plot that compares the maximum values of these data for the selected players in the regular season and playoffs is the end result.

Regular Season Performance

```python
# Regular Season

GOAT = pd.DataFrame(GOAT)

# Split the DataFrame into 'Regular Season' and 'Playoffs'
regular_season_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Regular Season']

# Specify the desired players
desired_players = ['Lebron James', 'Michael Jordan', 'Kobe Bryant']

# Filter the data for the desired players
filtered_data = regular_season_data[regular_season_data['Player Name'].isin(desired_players)]

# Specify the desired columns for the max values
max_columns = ['Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create a DataFrame with only the max values
max_values = filtered_data.groupby('Player Name')[max_columns].max().reset_index()

# Create a grouped bar plot for max values
ax = max_values.set_index('Player Name').plot(kind='bar', figsize=(10, 6), width=0.8,)
plt.title('Max Values for Regular Season Stats Grouped Bar Plot')
plt.xlabel('Player Name')
plt.ylabel('Stats Value', rotation=0, labelpad=15)
plt.legend(title='Statistics', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

Playoffs Performance

```python
# Playoffs

GOAT = pd.DataFrame(GOAT)

# Split the DataFrame into 'Regular Season' and 'Playoffs'
regular_season_data = GOAT[GOAT['Regular-Season/Playoff'] == 'Playoffs']

# Specify the desired players
desired_players = ['Lebron James', 'Michael Jordan', 'Kobe Bryant']

# Filter the data for the desired players
filtered_data = regular_season_data[regular_season_data['Player Name'].isin(desired_players)]

# Specify the desired columns for the max values
max_columns = ['Total Rebound Percentage', 'Total Assist Percentage', 'Total Steal Percentage', 'Total Block Percentage', 'Total Points Per Game']

# Create a DataFrame with only the max values
max_values = filtered_data.groupby('Player Name')[max_columns].max().reset_index()

# Create a grouped bar plot for max values
ax = max_values.set_index('Player Name').plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Max Values for Playoffs Stats Grouped Bar Plot')
plt.xlabel('Player Name')
# plt.ylabel('Max Values')
plt.legend(title='Statistics', bbox_to_anchor=(1, 1), loc='upper left')
plt.ylabel('Stats Value', rotation=0, labelpad=15)
plt.show()
```
